```r
# Step 1: Install necessary libraries in R
install.packages("ggplot2")
install.packages("readr")
install.packages("dplyr")

# Step 2: Load required libraries
library(ggplot2)
library(readr)
library(dplyr)

# Step 3: Upload file using Python (Google Colab workaround)
system("pip install -q gdown")  # Ensure gdown is installed
system("gdown --id FILE_ID_HERE -O /content/spotify_data.csv")

# Step 4: Load the dataset
dataset <- read_csv('/content/spotify_data.csv')

# Step 5: Display a preview of the dataset
print(head(dataset))

# Step 6: Histogram for 'Streams'
ggplot(dataset, aes(x = Streams)) +
  geom_histogram(binwidth = 50000000, fill = 'skyblue', color = 'black') +
  labs(title = "Histogram of Streams", x = "Streams", y = "Frequency")

# Step 7: Box plot for 'Daily' stream counts
ggplot(dataset, aes(y = Daily)) +
  geom_boxplot(fill = 'lightgreen') +
  labs(title = "Box Plot of Daily Streams", y = "Daily Streams")

# Step 8: Scatter plot between 'Streams' and 'Daily'
ggplot(dataset, aes(x = Streams, y = Daily)) +
  geom_point(color = 'blue') +
  labs(title = "Scatter Plot of Streams vs Daily Streams",
       x = "Streams", y = "Daily Streams")
```
